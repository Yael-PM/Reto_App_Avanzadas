# Detección de Anorexia con Natural Lenguage Inference y Algoritmos Genéticos en Redes Sociales
## 1. Introducción
El análisis de microtextos en redes sociales como X (Twitter) para la detección temprana de Trastornos de la Conducta Alimentaria (TCA), específicamente la Anorexia Nerviosa, plantea un desafío metodológico de dos frentes: la volatilidad lingüística del usuario digital y la escasez de datos clínicos etiquetados. Los enfoques convencionales basados en la frecuencia de términos (como n-gramas o TF-IDF) resultan insuficientes al ignorar la semántica profunda, el sarcasmo y las sutilezas contextuales de las patologías psíquicas. Por otro lado, la aplicación directa de un ajuste fino (fine-tuning) sobre Modelos de Lenguaje Masivos (LLMs) como BETO expone al sistema a un sobreajuste (overfitting) severo debido a la reducida escala de los datasets disponibles, provocando que la red memorice el ruido léxico en lugar de generalizar el constructo clínico.

Para resolver esta encrucijada, esta investigación justifica técnicamente el desarrollo de una arquitectura híbrida en cascada orientada a la Evidencia 3. El sistema propuesto trasciende el uso aislado de modelos de lenguaje mediante la convergencia de dos ramas de la Inteligencia Artificial: la inferencia profunda y la heurística evolutiva. Al integrar la Inferencia de Lenguaje Natural (NLI) con Algoritmos Genéticos (AG), esta solución no solo busca maximizar de forma abstracta la precisión del sistema, sino formalizar, validar y optimizar un pipeline robusto y eficiente que simule el proceso de filtrado clínico primario, demostrando la madurez técnica y la riqueza pedagógica necesarias en la formación de la ingeniería de sistemas complejos.

## 2. Descripción del Modelo de Solución Propuesto
El modelo propuesto se despliega como un clasificador en cascada de dos etapas principales, donde la salida probabilística y densa de un Transformer de última generación es refinada y filtrada heurísticamente por un proceso evolutivo antes de la toma de decisiones del clasificador final. Todo el flujo se consolida de forma nativa a través de las siguientes etapas constitutivas:

1. **Generación de Embeddings mediante NLI (Etapa Profunda):** Cada microtexto limpio actúa como una premisa dentro de un entorno Zero-shot. Utilizando el modelo DeBERTa-v3 (symanto/sn-xlm-roberta-base-snli-mnli-anli-xnli), el texto se contrasta contra un banco de hipótesis sintéticas estructuradas mediante NotebookLM a partir de los criterios del DSM-5 (restricción, pánico al peso, distorsión de la imagen) y dimensiones transdiagnósticas del RDoC. La red calcula la probabilidad de implicación lógica (Entailment), abstrayendo el significado conceptual sin alterar los pesos internos del Transformer.

2. **Optimización de Vocabulario mediante Algoritmos Genéticos (Etapa Heurística):** El NLI puro asume una importancia homogénea para todas las premisas médicas y genera vectores densos que pueden arrastrar redundancia o ruido contextual (como se observó con variables de control que sesgan la atención del modelo en el anterior alcance que solo usa NLI). Los embeddings generados se introducen en un Algoritmo Genético (AG) implementado con pygad. El cromosoma del AG codifica la activación/desactivación y el peso matemático de este "vocabulario de características semánticas". A través de operadores evolutivos (selección, cruce y mutación), la heurística descubre de manera automatizada el subconjunto óptimo de variables que maximiza el F1-Score balanceado, purgando los rasgos irrelevantes.

3. **Entrenamiento del Clasificador en Cascada (Etapa Supervisada):** El espacio de características purificado y ponderado por el AG alimenta la fase final de la cascada. Aquí se entrenan y evalúan competitivamente tres arquitecturas fundamentales: Random Forest (RF), Regresión Logística (LR) y el modelo de Naïve Bayes (NB). Esta cascada opera bajo un esquema de resiliencia lingüística: si la abstracción conceptual del Transformer decae ante la ironía o modismos de las redes sociales, la estructuración evolutiva y el clasificador downstream fungen como una red de seguridad estadística que corrige la predicción con base en las regularidades consolidadas del ecosistema de datos.

## Procesamiento del Dataset y Limpieza de Ruido
Los datos extraídos de redes sociales presentan ruido inherente, desde problemas de codificación (Mojibake) hasta terminología no estandarizada, lo que requiere un preprocesamiento cuidadoso para evitar sesgos en la interpretación semántica. Para estandarizar el corpus, se implementó un pipeline secuencial unificado:

1. **Restauración de Codificación:** Se aplicó la librería ftfy para corregir de manera prioritaria caracteres corruptos o mal codificados, garantizando la recuperación de tildes y eñes fundamentales en la semántica del español.

2. **Limpieza Estructural y Segmentación:** Mediante Expresiones Regulares (REGEX), se eliminaron URLs y menciones que carecen de valor clínico. Posteriormente, se aislaron hashtags compuestos (ej. "#hastaloshuesos") y se segmentaron semánticamente, a la par que se normalizó la jerga de internet (slang) hacia un español neutro. Esto previene que el modelo trate variaciones ortográficas de un mismo concepto como entidades desconectadas.

3. **Lematización Contextual:** Con la ayuda del modelo de spaCy, se redujeron las palabras a su raíz léxica (lemas). A diferencia de enfoques clásicos, se tomó la decisión clínica de no eliminar las "stopwords" (artículos, pronombres y conectores). La literatura médica reciente (como los estudios presentados en ACL Anthology) demuestra que la disminución en el uso de pronombres inclusivos ("nosotros") y conectores causales ("porque") actúan como biomarcadores de aislamiento social e indefensión aprendida. Eliminar estas palabras funcionales significaría la pérdida de señales predictivas críticas para la detección de perfiles de riesgo.

```python
import pandas as pd
import ftfy
import re
import spacy
from wordsegment import load, segment

# Inicialización de Modelos y Diccionarios ---
load() # Cargar wordsegment
nlp = spacy.load("es_core_news_sm")

dict_slang = {
    "ntp": "no te preocupes", "pq": "porque", "x q": "porque", 
    "tkm": "te quiero mucho", "d": "de", "q": "que", "toy": "estoy", 
    "chi": "si", "x": "por", "k": "que", "salu2": "saludos", 
    "bno": "bueno", "ño": "no", "onde": "donde", "tmb": "también", 
    "akiora": "a que hora", "aquiora": "a que hora", "shi": "si", 
    "xq": "porque", "xk": "porque"
}

# Nombres de Archivos
ARCHIVO_ENTRENAMIENTO = 'data_train.xlsx'
ARCHIVO_TEST          = 'data_test.csv'

def limpiar_pipeline_completo(texto):
    """
    Ejecuta todo el flujo de limpieza: rescate de caracteres, limpieza regex,
    normalización de slangs/hashtags y lematización.
    """
    if pd.isna(texto) or not isinstance(texto, str):
        return ""
    
    # 1. Rescate ftfy y Limpieza Básica (Regex)
    texto = ftfy.fix_text(texto)
    texto = re.sub(r'https?://\S+|www\.\S+', '', texto) # Quitar URLs
    texto = re.sub(r'@\w+', '', texto)                  # Quitar menciones
    texto = re.sub(r'[^\w\s.,!?¡¿#]', '', texto)        # Mantener puntuación básica y '#' para procesar después
    texto = texto.replace('\n', ' ').replace('\r', '')
    texto = re.sub(r'\s+', ' ', texto).strip().lower()  # Espacios extra y minúsculas
    
    # 2. Segmentación de Hashtags y Normalización de Slang
    palabras = texto.split()
    texto_procesado = []
    
    for p in palabras:
        p_limpia = p.replace('#', '')
        
        if p_limpia in dict_slang:
            texto_procesado.append(dict_slang[p_limpia])
        elif p.startswith('#') and len(p_limpia) > 5:
            seg = segment(p_limpia)
            if len(seg) < 4:
                texto_procesado.extend(seg)
            else:
                texto_procesado.append(p_limpia)
        else:
            texto_procesado.append(p_limpia) # Se inserta la palabra sin el '#'
            
    texto_unido = " ".join(texto_procesado)
    
    # 3. Lematización contextual con spaCy
    doc = nlp(texto_unido)
    lemas = [token.lemma_ for token in doc]
    
    return " ".join(lemas)


# Preparación de Variables para Entrenamiento
print("Cargando y procesando datasets (Train y Test)...")

# Carga de datos
df_train = pd.read_excel(ARCHIVO_ENTRENAMIENTO)
df_test = pd.read_csv(ARCHIVO_TEST, encoding='latin1')

# Aplicar el pipeline unificado
print("Aplicando pipeline de limpieza al set de Entrenamiento...")
df_train['tweet_procesado'] = df_train['tweet_text'].apply(limpiar_pipeline_completo)

print("Aplicando pipeline de limpieza al set de Prueba...")
df_test['tweet_procesado'] = df_test['tweet_text'].apply(limpiar_pipeline_completo)

# Guardado opcional
#df_train[['class', 'tweet_procesado']].to_csv('data_train_reparado.csv', index=False, encoding='utf-8')
#df_test[['class', 'tweet_procesado']].to_csv('data_test_reparado.csv', index=False, encoding='utf-8')

print("\n¡Procesamiento finalizado con éxito!")
print("\n##### Muestra Train #####")
print(df_train[['tweet_text', 'tweet_procesado']].head(2))
print("\n##### Muestra Test #####")
print(df_test[['tweet_text', 'tweet_procesado']].head(2))
```

## Extracción de Características Mediante Alineación Lógica-Semántica (NLI)

En las arquitecturas tradicionales de Machine Learning, la matriz de variables $X$ se construye contando la frecuencia de aparición de palabras (como en TF-IDF). En este nuevo enfoque de vanguardia, el paso a seguir consiste en **mecanizar un estándar médico internacional dentro de un Transformer**.

Para dotar al proyecto de un sustento científico, las variables de nuestra matriz $X$ se definieron mediante pares de hipótesis lógicas en un entorno de **Inferencia de Lenguaje Natural (NLI)**.

### El Rol de NotebookLM en la Ingeniería de Prompts Clínicos

Es fundamental destacar que el banco de hipótesis utilizado en este pipeline no fue diseñado de forma empírica o arbitraria. En su lugar, se empleó **NotebookLM** como una herramienta avanzada de curación de contenido y de ingeniería de prompts. A través de este entorno, se alimentaron fuentes y artículos indexados de alta confianza científica que describen de forma rigurosa tres marcos internacionales:

1. **Criterios Diagnósticos del DSM-5 (y CIE-11):** Mapeo directo de los criterios clínicos esenciales para la Anorexia Nerviosa: *Criterio A* (Restricción de la ingesta), *Criterio B* (Miedo intenso a ganar peso) y *Criterio C* (Distorsión de la imagen corporal).
2. **Dimensiones de Conducta RDoC (Research Domain Criteria del NIMH):** Incorporación de dimensiones psicológicas transversales y continuas, tales como los *Sistemas de Valencia Negativa* (culpa y autorreproche postsindromático) y *Sistemas de Regulación Cognitiva* (control obsesivo-rígido de rutinas).
3. **Biomarcadores Lingüísticos de Comunidades Web (Ciberpsicología):** Detección de la jerga de riesgo e idealización de los trastornos característica de entornos *Pro-Ana* y *Pro-Mia*.

NotebookLM extrajo la esencia de estas fuentes médicas y sintetizó el conocimiento en un conjunto de **frases asertivas, claras y descriptivas** (las cuales se detallan en el cuaderno de código adjunto a continuación). Este proceso garantiza que el modelo Transformer entienda el contexto cotidiano de redes sociales sin tecnicismos excesivos, pero manteniendo un anclaje clínico estricto.

### La Lógica de Contradicción y el Blindaje del Modelo

Para replicar el éxito de los sistemas de moderación por alineación de incrustaciones, el modelo necesita un **contraste de seguridad**. Por cada hipótesis de riesgo clínica formulada, se introduce una **hipótesis de control o neutralidad** (vía de escape semántica).

Al pasar el texto preprocesado (Premisa) y este banco de hipótesis sintetizado por NotebookLM por el codificador de **DeBERTa-v3** (específicamente entrenado en tareas XNLI), el modelo calcula la probabilidad de implicación lógica (*Entailment*). Si un tweet se alinea con una hipótesis de riesgo, su puntuación matemática en esa dimensión se eleva; si se alinea con una frase cotidiana o neutra, la puntuación disminuye.

El resultado de este paso será la transformación de texto crudo en matrices densas de probabilidades multidimensionales ($X_{train}$, $X_{val}$, $X_{test}$), listas para ser inyectadas en los clasificadores downstream (**Naïve Bayes, Regresión Logística y Random Forest**).

## Implementación del Pipeline Completo de División y Codificación NLI

El siguiente script en Python realiza la partición del dataset de entrenamiento (80% entrenamiento, 20% validación) para conservar la proporción del desbalance de clases. Posteriormente, utiliza **DeBERTa-v3** para mapear el banco de hipótesis generado por NotebookLM y extraer las matrices de características de forma eficiente.


```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Configuración del entorno y modelo transformer
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Utilizando dispositivo: {device}")

# Utilizaremos DeBERTa-v3-large optimizado para tareas de NLI en español/multilingüe
NLI_MODEL_NAME = "symanto/sn-xlm-roberta-base-snli-mnli-anli-xnli"
tokenizer = AutoTokenizer.from_pretrained(NLI_MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(NLI_MODEL_NAME).to(device)
model.eval()


# Definición del espacio de características clínicas y de seguridad (Hipótesis de riesgo vs. Hipótesis de control)
#    (Generadas por NotebookLM basándose en DSM-5, RDoC y Ciberpsicología)

# Hipótesis de riesgo (Positivas) extraídas de los criterios clínicos
hipotesis_riesgo = [
    "La persona describe una restricción deliberada de la ingesta calórica para lograr un peso significativamente bajo.",
    "El usuario manifiesta un temor persistente y paralizante a ganar peso o convertirse en obeso.",
    "El sujeto percibe su cuerpo o partes de él como excesivamente gordos a pesar de estar en infrapeso.",
    "El texto expresa sentimientos de culpa intensos o autorreproche extremo tras haber ingerido alimentos.",
    "La persona muestra un patrón de control obsesivo y reglas estrictas sobre horarios y cantidades exactas de comida.",
    "El mensaje promueve o idealiza la delgadez extrema como un objetivo de éxito y autovalía.",
    "El individuo menciona realizar ejercicio físico excesivo con el fin único de quemar calorías ingeridas.",
    "El usuario describe el chequeo constante de su peso en básculas o la medición repetitiva de sus dimensiones corporales.",
    "La persona niega o minimiza los peligros de su bajo peso actual y su malnutrición.",
    "El sujeto evita activamente situaciones sociales que involucren comida para ocultar sus hábitos restrictivos.",
    "El texto utiliza términos específicos como 'thinspo' o 'meanspo' para motivar el ayuno.",
    "El usuario describe comportamientos persistentes que sabotean cualquier intento médico o familiar de ganar peso.",
    "La persona se juzga a sí misma basándose exclusivamente en estándares de belleza externos e irreales.",
    "El mensaje revela una necesidad de control absoluto sobre el cuerpo para compensar otras áreas de la vida.",
    "El individuo utiliza fotografías (selfies) no publicadas para inspeccionar obsesivamente defectos corporales percibidos."
]

# Hipótesis de seguridad y escape semántico (Negativas) para mitigar falsos positivos
hipotesis_seguridad = [
    "La persona está comentando sobre el estado del tiempo o el clima en su ciudad.",
    "El texto discute características técnicas de un dispositivo electrónico o programa informático.",
    "El usuario describe tareas administrativas o de trabajo corporativo habituales.",
    "El mensaje trata sobre la salud, el juego o la alimentación equilibrada de un animal doméstico.",
    "La persona comparte una opinión o resumen sobre un libro de ficción que acaba de leer.",
    "El sujeto pide información sobre rutas de autobús o horarios de trenes locales.",
    "El texto describe planes para remodelar, pintar o limpiar una habitación de la casa.",
    "El usuario habla sobre el crecimiento de sus plantas o técnicas de siembra en el jardín.",
    "El mensaje comenta un acontecimiento político o económico global reciente.",
    "La persona menciona su preparación para un examen o la realización de un proyecto académico.",
    "El texto es una crítica neutral sobre la fotografía o dirección de una película.",
    "El individuo describe los pasos necesarios para renovar un documento de identidad.",
    "El usuario menciona datos geográficos o curiosidades sobre un país que desea visitar.",
    "El mensaje explica cómo armar un mueble o realizar una artesanía manual.",
    "El texto es un saludo cordial y sin contenido específico dirigido a un grupo de amigos."
]

# Consolidación del espacio de características extraído de NotebookLM
todas_las_hipotesis = hipotesis_riesgo + hipotesis_seguridad
NUM_CARACTERISTICAS = len(todas_las_hipotesis)

# Carga de datasets limpios y particiones de entrenamiento/validación
# Carga de los archivos limpios del pipeline anterior
df_train['tweet_procesado'] = df_train['tweet_procesado'].fillna("")
df_test['tweet_procesado']  = df_test['tweet_procesado'].fillna("")

# Clonamos la estructura para no alterar los DataFrames originales de tu celda previa
df_train_trabajo = df_train.copy()
df_test_trabajo  = df_test.copy()

# Realizamos la partición Estratificada: 80% Entrenamiento y 20% Validación
df_train_split, df_val_split = train_test_split(
    df_train_trabajo,
    test_size=0.20,
    random_state=42,
    stratify=df_train_trabajo['class'] # Mantiene el desbalance original de clases
)

print(f"Registros - Train: {len(df_train_split)} | Val: {len(df_val_split)} | Test: {len(df_test_trabajo)}")

# Extracción de embeddings semánticos vía NLI para cada conjunto (Train, Val, Test)
def generar_matriz_caracteristicas_nli(dataframe, columna_texto):
    """
    Pasa cada tweet preprocesado junto con el banco de hipótesis de NotebookLM por DeBERTa.
    Extrae la probabilidad de Entailment (Implicación) para construir la matriz X.
    """
    textos = dataframe[columna_texto].tolist()
    num_muestras = len(textos)
    
    # Inicializamos la matriz X con ceros
    X_matriz = np.zeros((num_muestras, NUM_CARACTERISTICAS))
    
    print(f"Codificando semánticamente {num_muestras} textos...")
    
    # Desactivamos el cálculo de gradientes (Inferencia Zero-Shot pura)
    with torch.no_grad():
        for idx, texto in enumerate(textos):
            if not texto.strip():
                continue # Deja la fila en ceros si el texto está completamente vacío
                
            # Duplicamos el texto de la premisa para emparejarlo con cada una de las hipótesis
            premisas = [texto] * NUM_CARACTERISTICAS
            
            # Tokenización en lote del par (Premisa, Hipótesis)
            inputs = tokenizer(
                premisas, 
                todas_las_hipotesis, 
                padding=True, 
                truncation=True, 
                return_tensors="pt"
            ).to(device)
            
            # Inferencia del modelo Transformer
            logits = model(**inputs).logits
            
            # Mapeo de salida clásico de Transformers NLI: índice 2 corresponde a 'Entailment'
            probs = torch.softmax(logits, dim=-1)
            entailment_scores = probs[:, 1].cpu().numpy()
            
            # Asignamos el vector resultante como la fila de características de la muestra
            X_matriz[idx] = entailment_scores
            
            if (idx + 1) % 100 == 0:
                print(f" -> Procesados {idx + 1}/{num_muestras} registros.")
                
    return X_matriz

# Generación final de matrices X e Y para downstream training
print("\n###### Extrayendo características semánticas para Entrenamiento (80%) ######")
X_train = generar_matriz_caracteristicas_nli(df_train_split, 'tweet_procesado')
y_train = df_train_split['class'].values

print("\n###### Extrayendo características semánticas para Validación (20%) ######")
X_val   = generar_matriz_caracteristicas_nli(df_val_split, 'tweet_procesado')
y_val   = df_val_split['class'].values

print("\n###### Extrayendo características semánticas para Prueba (Test Original) ######")
X_test  = generar_matriz_caracteristicas_nli(df_test_trabajo, 'tweet_procesado')
y_test  = df_test_trabajo['class'].values

print("\n[PROCESO TERMINADO] Matrices X e y generadas correctamente en memoria.")
print(f"Dimensiones finales de X_train: {X_train.shape}")
print(f"Cada renglón representa un tweet mapeado en {X_train.shape[1]} dimensiones probabilísticas clínico-conceptuales.")
```

## Descripción del Proceso de Optimización Evolutiva (En Cascada Semántica)

La implementación del Algoritmo Genético (AG) se rediseña para actuar como el filtro adaptativo central de nuestra arquitectura en cascada. En este nuevo enfoque, el cromosoma ya no representa palabras sueltas (n-gramas), sino que se define como una cadena binaria de 30 genes, donde cada gen corresponde a una de las dimensiones probabilísticas clínico-conceptuales extraídas previamente por DeBERTa-v3 (15 de riesgo DSM-5/RDoC y 15 de escape semántico). Si un gen toma el valor de uno, la hipótesis semántica se incluye en el vector; si es cero, se purga del sistema.

Para evaluar la viabilidad de cada combinación, la Fitness Function funge como el juez del proceso evolutivo operando bajo una estrategia de validación cruzada estratificada ($k=5$) exclusivamente sobre el conjunto de entrenamiento. Para elevar el rigor predictivo heremos un experimento, que consiste en que la función ejecuta internamente un barrido de hiperparámetros (GridSearch) sobre un clasificador base de Regresión Logística pero existe otra instancia que NO la aplica. Esto garantiza que la aptitud devuelta (métrica $F_1\text{-Score}$) refleje el rendimiento óptimo que ese subconjunto de características puede alcanzar, permitiendo al AG descubrir de forma automatizada la combinación exacta de intenciones clínicas que maximiza la detección, eliminando el ruido y blindando el pipeline contra el sarcasmo o sesgos del dataset de Twitter.

## Implementación de la Fitness Function Adaptada

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, f1_score

# 1. Definimos el scorer enfocado rigurosamente en la clase de riesgo psicológico
f1_anorexia = make_scorer(f1_score, pos_label='anorexia')

# Validación cruzada fija (k=5) para asegurar consistencia estadística en ambos experimentos
cv_estratificado = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =====================================================================
# EXPERIMENTO A: FITNESS FUNCTION SIN GRIDSEARCH INTERNO
# =====================================================================
def fitness_sin_grid(ga_instance, solution, solution_idx):
    """
    Evalúa la máscara binaria del AG usando una Regresión Logística estática.
    """
    sol = np.array(solution).flatten()
    
    # Penalización absoluta si el cromosoma apaga todo el vocabulario clínico
    if np.sum(sol) == 0:
        return 0.0
    
    try:
        # Filtramos la matriz X_train (30 columnas) usando la máscara del cromosoma
        X_reducida = X_train[:, sol == 1]
        
        # Clasificador downstream base (hiperparámetros por defecto)
        modelo_estatico = LogisticRegression(
            max_iter=1000, 
            class_weight='balanced', 
            solver='liblinear',
            random_state=42
        )
        
        # Validación cruzada interna para medir la aptitud general
        scores = cross_val_score(
            modelo_estatico, X_reducida, y_train, 
            cv=cv_estratificado, scoring=f1_anorexia, n_jobs=-1
        )
        fitness = scores.mean()
        
    except Exception as e:
        fitness = 0.0
        
    return fitness


# =====================================================================
# EXPERIMENTO B: FITNESS FUNCTION CON GRIDSEARCH INTERNO
# =====================================================================
def fitness_con_grid(ga_instance, solution, solution_idx):
    """
    Evalúa la máscara binaria buscando dinámicamente la mejor configuración 
    de regularización para el clasificador intermedio.
    """
    sol = np.array(solution).flatten()
    
    if np.sum(sol) == 0:
        return 0.0
    
    try:
        X_reducida = X_train[:, sol == 1]
        
        # Espacio de búsqueda acotado para el clasificador lineal
        param_grid = {'C': [0.01, 0.1, 1.0, 10.0]}
        
        # GridSearch interno que optimiza la Regresión Logística para esta combinación específica
        grid_search = GridSearchCV(
            LogisticRegression(max_iter=1000, class_weight='balanced', solver='liblinear', random_state=42),
            param_grid, cv=3, scoring=f1_anorexia, n_jobs=-1
        )
        
        scores = []
        # Ejecutamos el K-Fold manualmente para evaluar el rendimiento del GridSearch
        for train_idx, val_idx in cv_estratificado.split(X_reducida, y_train):
            X_tr, X_va = X_reducida[train_idx], X_reducida[val_idx]
            y_tr, y_va = y_train[train_idx], y_train[val_idx]
            
            grid_search.fit(X_tr, y_tr)
            # Extraemos la predicción utilizando el mejor estimador configurado
            mejor_modelo_interno = grid_search.best_estimator_
            preds = mejor_modelo_interno.predict(X_va)
            scores.append(f1_score(y_va, preds, pos_label='anorexia'))
            
        fitness = np.mean(scores)
        
    except Exception as e:
        fitness = 0.0
        
    return fitness
```

## Instanciación de parámetros del AG
Una vez declaradas las funciones de aptitud, procedemos a configurar las dos instancias de pygad.GA. El largo del cromosoma (num_genes) se acopla dinámicamente al tamaño actual de tu espacio de características extraído de DeBERTa (las 30 dimensiones que combinan los criterios DSM-5, RDoC y los escapes semánticos).

Mantendremos los hiperparámetros evolutivos idénticos en ambas instancias para asegurar que el experimento sea científicamente justo y que la única variable de cambio estructural sea la presencia del sub-GridSearch.

```python
import pygad

# Definimos el total de genes acoplado a las 30 hipótesis consolidadas en RAM
largo_cromosoma = X_train.shape[1]

# Configuración base del entorno evolutivo para ambos experimentos
configuracion_ag_base = {
    "num_generations": 30,         # Ciclos evolutivos balanceados para el tamaño del espacio (30 features)
    "num_parents_mating": 5,       # Cantidad de padres seleccionados para cruce
    "sol_per_pop": 15,             # Cromosomas activos por generación
    "num_genes": largo_cromosoma,
    "gene_space": [0, 1],          # Espacio binario estricto (0: Excluir, 1: Incluir hipótesis)
    "parent_selection_type": "sss",# Stochastic remainder selection
    "crossover_type": "single_point",
    "mutation_type": "random",
    "mutation_percent_genes": 10,  # Porcentaje de mutación adaptado para evitar estancamiento
    "random_seed": 42,
    "stop_criteria": ["reach_0.95"] # Criterio de parada temprana si alcanza rendimiento óptimo
}

# Instancia del Experimento 1: AG Tradicional con evaluación estática
ga_instancia_sin_grid = pygad.GA(
    fitness_func=fitness_sin_grid, 
    **configuracion_ag_base
)

# Instancia del Experimento 2: AG Híbrido con optimización interna de hiperparámetros
ga_instancia_con_grid = pygad.GA(
    fitness_func=fitness_con_grid, 
    **configuracion_ag_base
)

print(f"Instancias preparadas exitosamente para operar sobre matriz X con forma: {X_train.shape}")
```

## Integración de la Evolución y Evaluación de la Cascada

El núcleo de la arquitectura híbrida en cascada culmina con la ejecución simultánea de los dos experimentos evolutivos (Con y Sin GridSearch interno) y la posterior filtración de las características de DeBERTa en la memoria RAM. Al emplear el Algoritmo Genético, determinamos democráticamente qué conjunto de hipótesis clínicas y de control maximiza de manera estable el $F_1\text{-score}$ global mediante validación cruzada.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 1. Ejecución de la evolución para ambas instancias del AG
print("====== INICIANDO EXPERIMENTO 1: AG SIN GRIDSEARCH INTERNO ======")
ga_instancia_sin_grid.run()
sol_sin, fit_sin, _ = ga_instancia_sin_grid.best_solution()

print("\n====== INICIANDO EXPERIMENTO 2: AG CON GRIDSEARCH INTERNO ======")
ga_instancia_con_grid.run()
sol_con, fit_con, _ = ga_instancia_con_grid.best_solution()

# 2. Selección democrática de la mejor máscara evolutiva y análisis del vocabulario clínico seleccionado
print("\n" + "="*60)
print("             COMPARATIVA DE INSTANCIAS DEL AG")
print("="*60)
print(f"Mejor Fitness F1-Score (AG Sin Grid): {fit_sin:.4f} | Variables activas: {int(np.sum(sol_sin))}/30")
print(f"Mejor Fitness F1-Score (AG Con Grid): {fit_con:.4f} | Variables activas: {int(np.sum(sol_con))}/30")

if fit_con > fit_sin:
    cromosoma_ganador = np.array(sol_con).flatten()
    print("\nGanó la instancia CON GridSearch Interno.")
else:
    cromosoma_ganador = np.array(sol_sin).flatten()
    print("\nGanó la instancia SIN GridSearch Interno (o empataron).")

# Filtrado y visualización del vocabulario óptimo seleccionado por el AG
indices_optimos = np.where(cromosoma_ganador == 1)[0]
hipotesis_seleccionadas = [todas_las_hipotesis[i] for i in indices_optimos]

print(f"\n[VOCABULARIO ÓPTIMO] El AG ha purgado el ruido. {len(indices_optimos)} hipótesis conservadas:")
for idx, hip in zip(indices_optimos, hipotesis_seleccionadas):
    tipo = "Riesgo" if idx < 15 else "Control"
    print(f" -> [{tipo}] (Idx {idx}): {hip[:75]}...")

# Aplicamos el filtro de características directo en la RAM para el entrenamiento
X_train_opt = X_train[:, cromosoma_ganador == 1]
```

Una vez descubierto el cromosoma con el "vocabulario óptimo" más apto, procedemos a aislar esas columnas específicas sobre las matrices $X$ en la memoria. Inmediatamente después, el flujo alimenta la fase supervisada final, donde ponemos a competir de forma paralela tres filosofías algorítmicas: Naïve Bayes, Regresión Logística y Random Forest. Para asegurar el rigor científico del reporte de entrenamiento, la evaluación se procesa mediante un esquema unificado de Stratified K-Fold Cross-Validation ($k=5$) utilizando variables y figuras explícitas dentro de Matplotlib y Seaborn, garantizando un diagnóstico visual íntegro del pipeline antes de enfrentarse a los conjuntos de validación o prueba.

```python

# 3. Función de evaluación y graficación de modelos downstream con reportes detallados, matrices de confusión estilizadas y curvas ROC/AUC conjuntas
def evaluar_y_graficar_modelos(modelos, X, y, nombre_fase, usar_cv=False):
    """
    Genera reportes de clasificación, matrices de confusión elegantes con Seaborn
    y curvas ROC/AUC de forma agrupada para Naïve Bayes, LR y RF.
    """
    print(f"\n{'='*60}")
    print(f"EVALUACIÓN DOWNSTREAM: {nombre_fase.upper()}")
    print(f"{'='*60}")

    # Inicialización del lienzo para las matrices de confusión en fila
    fig_cm, axes_cm = plt.subplots(1, len(modelos), figsize=(6.5 * len(modelos), 4.5))
    if len(modelos) == 1: 
        axes_cm = [axes_cm]

    # Inicialización del lienzo para las curvas ROC conjuntas
    fig_roc, ax_roc = plt.subplots(figsize=(8, 5))
    sns.set_theme(style="whitegrid")
    
    for i, (nombre, modelo) in enumerate(modelos.items()):
        if usar_cv:
            # Implementación estricta de Stratified K-Fold (k=5) en el set de entrenamiento
            skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
            y_pred = cross_val_predict(modelo, X, y, cv=skf, n_jobs=-1)
            y_probas = cross_val_predict(modelo, X, y, cv=skf, method='predict_proba', n_jobs=-1)
            # Ajuste final sobre el set completo para su persistencia en RAM
            modelo.fit(X, y)
        else:
            y_pred = modelo.predict(X)
            y_probas = modelo.predict_proba(X)

        # Despliegue del reporte estadístico detallado
        print(f"\n--- Reporte de Clasificación: {nombre} ---")
        print(classification_report(y, y_pred, zero_division=0))

        # Matriz de Confusión estilizada en Seaborn (Ejes explícitos)
        cm = confusion_matrix(y, y_pred)
        sns.heatmap(
            cm, annot=True, fmt='d', cmap='Blues' if usar_cv else 'Greens', 
            ax=axes_cm[i], cbar=False, xticklabels=modelo.classes_, yticklabels=modelo.classes_,
            annot_kws={"size": 12, "weight": "bold"}
        )
        axes_cm[i].set_title(f"Confusión: {nombre}", fontsize=12, pad=10)
        axes_cm[i].set_xlabel("Predicción del Modelo", fontsize=10)
        axes_cm[i].set_ylabel("Clase Real (Golden Standard)", fontsize=10)

        # Cálculo de la curva ROC/AUC identificando dinámicamente el índice de 'anorexia'
        idx_anorexia = list(modelo.classes_).index('anorexia')
        y_real_bin = [1 if label == 'anorexia' else 0 for label in y]
        
        fpr, tpr, _ = roc_curve(y_real_bin, y_probas[:, idx_anorexia])
        auc_score = roc_auc_score(y_real_bin, y_probas[:, idx_anorexia])
        
        ax_roc.plot(fpr, tpr, label=f"{nombre} (AUC = {auc_score:.4f})", linewidth=2)

    # Configuración final del espacio cartesiano de las curvas ROC
    ax_roc.plot([0, 1], [0, 1], 'k--', alpha=0.5) 
    ax_roc.set_xlabel('Tasa de Falsos Positivos (1 - Especificidad)', fontsize=11)
    ax_roc.set_ylabel('Tasa de Verdaderos Positivos (Recall)', fontsize=11)
    ax_roc.set_title(f"Curvas ROC Conjuntas - {nombre_fase}", fontsize=13, pad=12)
    ax_roc.legend(loc="lower right", fontsize=10)

    plt.tight_layout()
    plt.show()

    return modelos
```

```python
# =====================================================================
# 4. INSTANCIACIÓN DE LOS CLASIFICADORES DOWNSTREAM Y EVALUACIÓN DE ENTRENAMIENTO
# =====================================================================
# Declaramos los tres algoritmos competidores con balanceo de pesos condicional
modelos_cascada_base = {
    "Naïve Bayes (Gaussian)": GaussianNB(),
    "Regresión Logística": LogisticRegression(max_iter=1000, class_weight='balanced', solver='liblinear', random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42, n_jobs=-1)
}

# Ejecución del pipeline con Validación Cruzada (usar_cv=True) sobre la matriz purificada
dict_modelos_entrenados = evaluar_y_graficar_modelos(
    modelos=modelos_cascada_base, 
    X=X_train_opt, 
    y=y_train, 
    nombre_fase="Entrenamiento (5-Fold Cross-Validation)", 
    usar_cv=True
)
```
