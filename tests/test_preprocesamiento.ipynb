```python
import unittest
import pandas as pd
import numpy as np
import ftfy
import re

# Función bajo prueba
def preprocesamiento(texto):
    if pd.isna(texto) or not isinstance(texto, str):
        return ""
    texto = ftfy.fix_text(texto)
    texto = re.sub(r'https?://\S+|www\.\S+', '', texto)
    texto = re.sub(r'@\w+', '', texto)
    texto = re.sub(r'[^\w\s.,!?¡¿]', '', texto)
    texto = texto.replace('\n', ' ').replace('\r', '')
    texto = re.sub(r'\s+', ' ', texto)
    return texto.strip().lower()
```

```python
import numpy as np
import spacy
from wordsegment import load, segment
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, f1_score

load()

dict_slang = {
    "ntp": "no te preocupes", "pq": "porque", "x q": "porque",
    "tkm": "te quiero mucho", "d": "de", "q": "que", "toy": "estoy",
    "chi": "si", "x": "por", "k": "que", "salu2": "saludos",
    "bno": "bueno", "ño": "no", "onde": "donde", "tmb": "también",
    "akiora": "a que hora", "aquiora": "a que hora", "shi": "si",
    "xq": "porque", "xk": "porque"
}

def segmentar_y_normalizar_v2(texto):
    if not texto: return ""
    palabras = texto.split()
    texto_procesado = []
    for p in palabras:
        p_limpia = p.replace('#', '').lower()
        if p_limpia in dict_slang:
            texto_procesado.append(dict_slang[p_limpia])
        elif p.startswith('#') and len(p_limpia) > 5:
            seg = segment(p_limpia)
            if len(seg) < 4:
                texto_procesado.extend(seg)
            else:
                texto_procesado.append(p_limpia)
        else:
            texto_procesado.append(p)
    return " ".join(texto_procesado)

_nlp = spacy.load("es_core_news_sm")

def tokenizar_y_lematizar(texto):
    if not texto: return ""
    doc = _nlp(texto)
    return " ".join(token.lemma_ for token in doc)

f1_anorexia = make_scorer(f1_score, pos_label='anorexia')

def fitness_function(ga_instance, solution, solution_idx):
    sol = np.array(solution).flatten()
    if np.sum(sol) == 0:
        return 0.0
    try:
        X_reducida = X[:, sol == 1]
        modelo = LogisticRegression(max_iter=1000, class_weight='balanced', solver='liblinear')
        scores = cross_val_score(modelo, X_reducida, y, cv=5, scoring=f1_anorexia)
        return float(scores.mean())
    except Exception:
        return 0.0
```

## Suite 10 — `segmentar_y_normalizar_v2`: expansión de slang

```python
class TestSegmentarSlang(unittest.TestCase):

    def test_xq_expandido(self):
        resultado = segmentar_y_normalizar_v2("xq no comiste")
        self.assertIn("porque", resultado)
        self.assertNotIn("xq", resultado)

    def test_tmb_expandido(self):
        resultado = segmentar_y_normalizar_v2("tmb tengo hambre")
        self.assertIn("también", resultado)
        self.assertNotIn("tmb", resultado)

    def test_tkm_expandido(self):
        resultado = segmentar_y_normalizar_v2("tkm")
        self.assertEqual(resultado, "te quiero mucho")

    def test_slang_insensible_a_mayusculas(self):
        # p_limpia aplica .lower(), así que "XQ" debe expandirse igual que "xq"
        resultado = segmentar_y_normalizar_v2("XQ lloras")
        self.assertIn("porque", resultado)

    def test_salu2_expandido(self):
        resultado = segmentar_y_normalizar_v2("salu2 a todos")
        self.assertIn("saludos", resultado)

unittest.main(argv=[''], exit=False, verbosity=2)
```

## Suite 11 — `segmentar_y_normalizar_v2`: segmentación de hashtags

```python
class TestSegmentarHashtag(unittest.TestCase):

    def test_hashtag_largo_produce_multiples_tokens(self):
        # "#hastaloshuesos" (14 chars) debe segmentarse en varias palabras
        resultado = segmentar_y_normalizar_v2("#hastaloshuesos")
        self.assertNotIn("#", resultado)
        self.assertGreater(len(resultado.split()), 1)

    def test_simbolo_hashtag_eliminado(self):
        resultado = segmentar_y_normalizar_v2("#proanorexia")
        self.assertNotIn("#", resultado)

    def test_hashtag_mayusculas_normalizado(self):
        # "#HastaLosHuesos" → segmentación en minúsculas sin #
        resultado = segmentar_y_normalizar_v2("#HastaLosHuesos")
        self.assertNotIn("#", resultado)
        self.assertEqual(resultado, resultado.lower())

    def test_hashtag_corto_no_segmentado(self):
        # "#ana" (3 chars) ≤ 5, no entra en la rama de segmentación
        resultado = segmentar_y_normalizar_v2("#ana")
        self.assertIn("ana", resultado)

unittest.main(argv=[''], exit=False, verbosity=2)
```
